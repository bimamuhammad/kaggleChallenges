# Build Data Loading Function

## Define Paths to files

```python
CAPUCHIN_FILE = os.path.join('Parsed_Capuchinbird_Clips', 'XC3776-3.wav')
NOT_CAPUCHIN_FILE = os.path.join('Parsed_Not_Capuchinbird_Clips', 'afternoon-birds-song-in-forest-0.wav')
```

## Build Loading Function

```python
def load_wav_16k_mono(filename):
  # load encoded wav file
  file_content = tf.io.read_file(filename)
  #Decode wav (tensor by channels)
  wav, sample_rate = tf.audio.decode_wav(file_content, desired_channels=1)
  #Remove trailing axis
  wav = tf.squeeze(wav, axis=-1)
  sample_rate = tf.cast(sample_rate, dtype=tf.int64)
  #Goes from 44100Hz to 16000hz - amplitude of audion signal
  wav = tfio.audio.resample(wav, rate_in=sample_rate, rate_out=16000)
  return wav
```

## Create TensorFlow Dataset

### Define Paths to Positive and Negative Data

```python
POS = os.path.join('Parsed_Capuchinbird_Clips')
NEG = os.path.join('Parsed_Not_Capuchinbird_Clips')
```

## Create Tensorflow datasets

```python
pos = tf.data.Dataset.list_files(POS+'\*.wav')
neg = tf.data.Dataset.list_files(NEG+'\*.wav')
```

## Add labels AND COMBINE POSITIVE AND NEGATIVE SAMPLES

```python
positives = tf.data.Dataset.zip((pos, tf.data.Dataset.from_tensor_slices(tf.ones(len(pos)))))
negatives = tf.data.Dataset.zip((neg, tf.data.Dataset.from_tensor_slices(tf.zeros(len(neg)))))
data = positives.concatenate(negatives)
```

# Build Preprocessing FUnction to convert to spectrogram

## Build Preprocessing function

```python
def preprocess(file_path, label):
    wav = load_wav_16k_mono(file_path)
    wav = wav[:48000]
    zero_padding = tf.zeros([48000] - tf.shape(wav), dtype=tf.float32)
    wav = tf.concat([zero_padding, wav], 0)
    spectrogram = tf.signal.stft(wav, frame_length=320, frame_step=32)
    spectrogram = tf.abs(spectrogram)
    spectrogram = tf.expand_dims(spectrogram, axis=2)
    return spectrogram, label
```

## Test Out the function and Viz the Spectrogram

```python
filepath, label = negatives.shuffle(buffer_size=10000).as_numpy_iterator().next()
```

```python
spectrogram, label = preprocess(filepath, label)
```

```python
plt.figure(figsize=(30,20))
plt.imshow(tf.transpose(spectrogram)[0])
plt.show()
```

# Create Training and Testing Partitions

## Create a tensorflow data pipeline

```python
data = data.map(preprocess)
data = data.cache()
data = data.shuffle(buffer_size=1000)
data = data.batch(16)
data = data.prefetch(8)
```

## Split into Traning and Testing Partition

```python
train = data.take(36)
test = data.skip(36).take(15)
```

## Test One Batch

```python
samples, labels = train.as_numpy_iterator().next()
```

# Build Deep Leanring Model

## Load Tensorflow Dependencies

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Flatten
```

## Build Sequential Model, Compile and View Summary

```python
model = Sequential()
model.add(Conv2D(16, (3,3), activation='relu', input_shape=(1491, 257,1)))
model.add(Conv2D(16, (3,3), activation='relu'))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('Adam', loss='BinaryCrossentropy', metrics=[tf.keras.metrics.Recall(), tf.keras.metrics.Precision()])
```

```python
model.summary()
```

## Fit Model, View Loss and KPI Plots

```python
hist = model.fit(train, epochs=4, validation_data=test)
```

```python
plt.title('Loss')
plt.plot(hist.history['loss'], 'r')
plt.plot(hist.history['val_loss'], 'b')
plt.show()
```

```python
plt.title('Precision')
plt.plot(hist.history['precision'], 'r')
plt.plot(hist.history['val_precision'], 'b')
plt.show()
```

```python
plt.title('Recall')
plt.plot(hist.history['recall'], 'r')
plt.plot(hist.history['val_recall'], 'b')
plt.show()
```

# Build forest Parsing Function

## Load up MP3s

```python
def load_mp3_16k_mono(filename):
    """ Load a WAV file, convert it to a float tensor, reample to 16 kHz single channel """
    res = tfio.audio.AudioIOTensor(filename)
    #Convert to tensor and combine channels
    tensor = res.to_tensor()
    tensor = tf.math.reduce_sum(tensor, axis=1) / 2
    
    #Extract sample rate and cast
    sample_rate = res.rate
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    
    #Resample to 16 kHz
    wav = tfio.audio.resample(tensor, rate_in=sample_rate, rate_out=16000)
    return wav
```
